```python
data=pd.read_csv('C:/Users/basleal/Desktop/tutorial/kaim-week-3/kaim-week-3/data/raw/MachineLearningRating_v3.csv')
data.head()
```

```python
for column in data.columns:
    missing_count = data[column].isnull().sum()
    if missing_count>0:
        print(column, missing_count)
```

```python
data.drop(columns=['Unnamed: 0','Bank','AccountType','CrossBorder', 'NumberOfVehiclesInFleet', 'WrittenOff', 'Rebuilt', 'Converted'],inplace=True)
```

```python
map_cat={'Mr':'Male','Mrs':'Female','Miss':'Female','Ms':'Female','Dr':'Male'}
data['Gender']=data['Title'].map(map_cat)
```

```python
vis_columns=[]
for column in data.columns:
    
    missing_count = data[column].isnull().sum()
    if missing_count>0:
        vis_columns.append(column)
        print(column, missing_count)
print(vis_columns)
```

```python
print(data['MaritalStatus'].value_counts())
map_mar={'Mrs':'married','Miss':'Single','Ms':'Single'}
data['MaritalStatus']=data['MaritalStatus'].fillna(data['MaritalStatus'].map(map_mar))
data['MaritalStatus'].isna().sum()

```

```python
categorical=[]
numerical=[]
for column in data.columns:
    if data[column].dtypes=='object':
        categorical.append(column)
    elif data[column].dtypes in ['int64','float64']:
        numerical.append(column)
print('categorical',categorical,'\n','numerical',numerical)
```

```python

data['MaritalStatus']=data['MaritalStatus'].fillna(data['MaritalStatus'].mode()[0])
data['MaritalStatus'].isna().sum()
```

```python
for column in data.columns:
    if column in categorical:
        data[column].fillna(data[column].mode()[0],inplace=True)
    elif column in numerical:
        data[column].fillna(data[column].mean(),inplace=True)
data.isnull().sum()
```

```python
data.to_csv('cleanedMachineLearningRating_v3.csv',index=False)
```
