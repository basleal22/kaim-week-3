```python
from sklearn.preprocessing import LabelEncoder
#initialize
Label_encoder=LabelEncoder()
#list of columns to encode
Label_columns=['MaritalStatus', 'NewVehicle', 'ExcessSelected','IsVATRegistered']
onehot_columns = ['Citizenship', 'LegalType', 'Title', 'Language', 'Gender', 'Country', 'Province', 
                  'MainCrestaZone', 'SubCrestaZone', 'ItemType', 'VehicleType', 'make', 'Model', 
                  'bodytype', 'AlarmImmobiliser', 'TrackingDevice', 'TermFrequency', 'CoverCategory', 
                  'CoverType', 'CoverGroup', 'Section', 'Product', 'StatutoryClass', 'StatutoryRiskType']
```

```python
#apply label encoding for ordinal columns
for column in Label_columns:
    data[column]=Label_encoder.fit_transform(data[column])
#apply One-Hot Encoding for nominal columns
data=pd.get_dummies(data,columns=onehot_columns,drop_first=True)
data.head()
```
